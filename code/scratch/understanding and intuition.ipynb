```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib.patches import FancyArrowPatch
```

```python
def func(x):
    return x*x + 2

def func_prime(x):
    return 2*x
```

# Plot the function


```python
x_axis = np.arange(-10, 10, 0.5)
y_axis = func(x_axis)
```

```python
plt.plot(x_axis, y_axis)
```

```python
def slope_func_at_a(func,x, h = 0.0001):
    return (func(x + h) - func(x))/h
    
```

```python
slopes = func_prime(x_axis)
```

```python
def drawArrow(ax, from_, to_, scale=20):
    arrow = FancyArrowPatch(from_, to_, mutation_scale=scale, color='green')
    ax.add_patch(arrow)
```

```python
fig, ax = plt.subplots(figsize=(100, 100))
ax.plot(x_axis, y_axis)
l = 5
mm = max(slopes)
mmg = 100
sign = lambda x : abs(x) / x if not x == 0 else x
for i, slope in enumerate(slopes):
    ssl = sign(slope)
    mid = (x_axis[i] + ssl * 0.5, y_axis[i] + ssl * 0.5)
    aslope = math.atan(slope)
    cs = math.cos(aslope)
    ss = math.sin(aslope)
    from_ = (mid[0] - l * cs, mid[1] - l * ss)
    to_ = (mid[0]   + l * cs, mid[1] + l * ss)
    if ssl > 0:
        drawArrow(ax, from_, to_, int(mmg * slope/mm))
    else:
        drawArrow(ax, to_, from_, int(mmg * slope/mm))
```
