```python
from skimage.feature import graycomatrix, graycoprops
img1_a = np.array(img2, dtype=np.uint8)
glcm = graycomatrix(img1_a, distances=[3], angles=[45], levels=256)
```

```python
graycoprops(glcm, "contrast")
```

```python
def getprop(img, prop):
    img1_a = np.array(img, dtype=np.uint8)
    glcm = graycomatrix(img1_a, distances=[3], angles=[45], levels=256)
    return graycoprops(glcm, prop).reshape(-1, )[0]
    
```

```python
getprop(img1, "contrast")
```
